# 📂 1. Load and Inspect the Data

## Loading The Data

```python
import pandas as pd
cdi_df = pd.read_csv("U.S._Chronic_Disease_Indicators_20250516.csv")
cdi_df
```

## Basic overview

```python
print("Dataset Shape:", cdi_df.shape)
print("\nDataset Info:\n", cdi_df.info())
print("\nFirst Few Rows:\n", cdi_df.head())
```

# 🛠️ 2. Handle Missing Values

## How Many Nulls 

```python
# Check missing values
print("Nulls:\n", cdi_df.isnull().sum())
```

```python
# Check percentage of missing values
missing_percentage = (cdi_df.isnull().sum() / len(cdi_df)) * 100
print("Nulls Percentage:\n", missing_percentage)
```

## Let's Clean

```python
print("Available Columns:", cdi_df.columns)
```

```python
# Drop Completely Empty Columns
cdi_df.drop(columns=['Response', 'StratificationCategory2', 'Stratification2',
                     'StratificationCategory3', 'Stratification3', 'ResponseID', 'StratificationCategoryID2', 'StratificationID2', 'StratificationCategoryID3',
                    'StratificationID3'],
            inplace=True, )
```

```python
print("Available Columns:", cdi_df.columns)
```

```python
# Handle Missing Values in Numeric Columns
# I chose to use median imputation, as it is robust against outliers
num_cols = ['DataValue', 'DataValueAlt', 'LowConfidenceLimit', 'HighConfidenceLimit']
for col in num_cols:
    cdi_df[col].fillna(cdi_df[col].median(), inplace=True)
```

```python
# Drop Non-Essential Text-Based Columns
cdi_df.drop(columns=['DataValueFootnoteSymbol', 'DataValueFootnote'], inplace=True)
```

```python
print("Remaining Missing Values:\n", cdi_df.isnull().sum())
```

# 🎭 3. Data Formatting and Standardization

```python
# Converting column names to lowercase for consistency
cdi_df.columns = cdi_df.columns.str.lower()
```

```python
# Standardize categorical values
cdi_df['locationdesc'] = cdi_df['locationdesc'].str.strip().str.title()
```

# 📊 4. Summary Statistics & Exploration

```python
print("Summary Statistics:\n", cdi_df.describe())
```

```python
# Unique locations and topics
print("Number of Locations:", cdi_df['locationdesc'].nunique())
print("Topics Covered:\n", cdi_df['topic'].unique())
```

# 📈 5. Visualizing Trends

```python
import matplotlib.pyplot as plt
```

## 📊1️⃣ Bar Chart for Chronic Disease Prevalence by State

```python
# Filter out the aggregate national-level data
filtered_df = cdi_df[cdi_df['locationdesc'].str.lower() != 'united states']
# Group data by location and calculate the mean prevalence
state_avg = filtered_df.groupby('locationdesc')['datavalue'].mean().sort_values(ascending=False).head(15)
# Create a horizontal bar plot
plt.figure(figsize=(12, 6))
plt.barh(state_avg.index, state_avg.values, color='steelblue')
plt.xlabel("Average Disease Prevalence (%)")
plt.ylabel("State")
plt.title("Top 15 States with Highest Chronic Disease Prevalence")
plt.gca().invert_yaxis()  
plt.show()
```

## 📈 2️⃣ Line Chart for Trends Over Time

```python
# Aggregate data by year
yearly_trends = cdi_df.groupby('yearstart')['datavalue'].mean()

# Plot line graph
plt.figure(figsize=(10, 5))
plt.plot(yearly_trends.index, yearly_trends.values, marker='o', linestyle='-', color='blue')
plt.title("Average Chronic Disease Prevalence Over Time")
plt.xlabel("Year")
plt.ylabel("Average Prevalence Rate")
plt.xticks(yearly_trends.index, rotation=45)  
plt.grid(True)
plt.show()
```

## 🔄 3️⃣ Histogram for Disease Distribution

```python
import numpy as np
# Check min and max 
min_val = cdi_df.datavalue.min()
max_val = cdi_df.datavalue.max()
print("Min:", min_val, "Max:", max_val)
# used numpy to adjust bins
bins = np.linspace(min_val, max_val, 4)
plt.figure(figsize=(10, 5))
plt.hist(cdi_df.datavalue, bins=bins, color='green', edgecolor='black')
plt.title("Distribution of Chronic Disease Indicators")
plt.xlabel("Data Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 📊4️⃣ Bar Chart for Average Chronic Disease Indicator By Topic

```python
# Group the data by topic and compute the mean of 'datavalue' for each topic
topic_avg = cdi_df.groupby('topic')['datavalue'].mean().sort_values(ascending=True).tail(5)
plt.figure(figsize=(10, 6))
plt.barh(topic_avg.index, topic_avg.values, color='skyblue')
plt.xlabel("average data value")
plt.ylabel("topic")
plt.title("top 5 average chronic disease indicator by topic")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 💹5️⃣ Scatter Plot for Confidence Limit Impact

```python
plt.figure(figsize=(10, 6))
plt.scatter(cdi_df['lowconfidencelimit'], cdi_df['datavalue'], s=30, alpha=0.5, color='purple')
plt.title("chronic disease prevalence vs. low confidence limit")
plt.xlabel("low confidence limit")
plt.ylabel("chronic disease prevalence (%)")
plt.grid(True)
plt.show()
```
